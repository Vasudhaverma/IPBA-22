```python
import pandas as pd
import os
import datetime
import snscrape.modules.twitter as sntwitter
```

```python
startDate = datetime.date(2020,1,30)
untilDate = startDate + datetime.timedelta(days=1)
currDate = datetime.date.today()
# currDate = datetime.date(2020,2,1)

dateList = []


def daterange(startDate, endDate):
#   return [start + datetime.timedelta(n) for n in range(int((end - start).days))]
    for n in range(int ((endDate - startDate).days)+1):
        yield startDate + datetime.timedelta(n)

for dt in daterange(startDate, currDate):
    dateList.append(dt)
#     print(dt.strftime("%Y-%m-%d"))

# print(dateList)
```

```python
maxTweets = 500
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for sdate in dateList:
    os.chdir('/Users/vasudhaverma/Desktop/IPBA')
    untilDate = sdate + datetime.timedelta(days=1)
    #Keyword: Coronavirus
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Coronavirus since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #keyword:Covid
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Covid since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:Covid19
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Covid19 since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:Novel Coronavirus
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Novel Coronavirus since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:PPE Kits
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'PPE Kits since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:Mask
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Mask since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:Vaccine
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Vaccine since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:Vaccination
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Vaccination since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:CovidShield
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'CovidShield since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:Covaxin
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Covaxin since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:Sputnik
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Sputnik since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:Pfizer
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Pfizer since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    #Keyword:Moderna
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'Moderna since:{sdate:%Y-%m-%d} until:{untilDate:%Y-%m-%d}').get_items()):
        if i>=maxTweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.location])
    
    # Creating a dataframe from the tweets list above
    tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Location'])
    
    #Emptying tweetslist:
    tweets_list1 = []
    
    # Export dataframe into a CSV
    os.chdir('/Users/vasudhaverma/Desktop/IPBA/TweetsDataset')
    file_name = 'tweet'+sdate.strftime("%Y%m%d")+'.csv'
    tweets_df1.to_csv(file_name, sep=',',)
```
